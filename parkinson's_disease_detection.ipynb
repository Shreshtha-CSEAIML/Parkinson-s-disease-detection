```python
#importing dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

data collection and analysis

```python
#loading the data from csv file to a pandas dataframe
parkinson_data  = pd.read_csv('/content/parkinsons.csv')
```

data pre-processing

```python
import warnings
warnings.filterwarnings("ignore", message="X does not have valid feature names, but StandardScaler was fitted with feature names")
X = parkinson_data.drop(columns = ['name','status'], axis=1)
Y = parkinson_data['status']
```

```python
print(X)
```

```python
print(Y)
```

splitting the data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

data standardisation

```python
scalar = StandardScaler()
```

```python
scalar.fit(X_train)
```

```python
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(parkinson_data.drop('name', axis=1).corr(), annot=False, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
print(X_train)
```

model training

support vector machine model

```python
model = svm.SVC(kernel='linear')
```

```python
# training the SVM model with training data
model.fit(X_train, Y_train)
```

model evaluation

accuracy score

```python
#accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('accuracy score of training data : ',training_data_accuracy  )
```

```python
#accuracy score on testing data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('accuracy score of test data : ',test_data_accuracy  )
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, X_test_prediction)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

building a predictive system

```python
input_data = (162.56800, 198.34600,77.63000,0.00502,0.00003,0.00280,0.00253,0.00841,0.01791,0.16800,0.00793,0.01057,0.01799,0.02380,0.01170,25.67800,0.427785,0.723797,-6.635729,0.209866,1.957961,0.135242
)
#changing the input data into numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#standardize the data
std_data = scalar.transform(input_data_reshaped)

prediction = model.predict(std_data)
print(prediction)

if(prediction[0] == 0 ):
  print('the person does not have parkinson disease')
else:
  print('the person has parkinson disease')
```

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(Y_test, model.decision_function(X_test))
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label='ROC Curve (area = %0.2f)' % roc_auc)
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
